```python
import pandas as pd
```

```python
layoffs_data = pd.read_csv('/content/layoffs_data.csv')
layoffs_data.head()
```

```python
location_based_df_country = layoffs_data[['Laid_Off_Count','Country']].groupby('Country').sum()
location_based_df_country.head()
```

```python
location_based_df_country = location_based_df_country.sort_values(by=['Laid_Off_Count'],ascending=False)
location_based_df_country.head()
```

```python
percentage_country_df = location_based_df_country.reset_index()
percentage_country_df.head()
```

```python
percentage_country_df['Laid Off Percentage'] = (percentage_country_df['Laid_Off_Count']/percentage_country_df['Laid_Off_Count'].sum()*100).round(decimals=2)
percentage_country_df.head()
```

```python
len(percentage_country_df)
```

```python
pie_chart_df = percentage_country_df.head(10)
pie_chart_df
```

```python
percentage_country_df.tail(47)
```

```python
percentage_country_df['Laid_Off_Count'].tail(47).sum()
```

```python
other_ls = {'Country':'Other','Laid_Off_Count':percentage_country_df['Laid_Off_Count'].tail(47).sum(),'Laid Off Percentage':(percentage_country_df['Laid_Off_Count'].tail(47).sum()/percentage_country_df['Laid_Off_Count'].sum()*100).round(decimals=2)}
other_ls
```

```python
pie_chart_df = pie_chart_df.append(other_ls,ignore_index=True)
pie_chart_df
```

```python
pie_chart_df = pie_chart_df.set_index('Country')
pie_chart_df
```

```python
explode = (0.02, 0.07, 0.06, 0.05, 0.03, 0.028, 0.026, 0.024, 0.022, 0.02, 0.01)
pie_chart_df.plot.pie(y='Laid Off Percentage', figsize = (10,10),autopct='%1.0f%%',explode=explode)
```

```python
industry_based_df = layoffs_data[['Laid_Off_Count','Industry']].groupby('Industry').sum()
industry_based_df.head()
```

```python
industry_based_df = industry_based_df.sort_values(by=['Laid_Off_Count'],ascending=False)
industry_based_df.head()
```

```python
industry_based_df.plot.bar()
```

```python
location_based_df_city = layoffs_data[['Laid_Off_Count','Location_HQ']].groupby('Location_HQ').sum()
location_based_df_city = location_based_df_city.sort_values(by=['Laid_Off_Count'],ascending=False)
percentage_city_df = location_based_df_city.reset_index()
percentage_city_df['Laid Off Percentage'] = (percentage_city_df['Laid_Off_Count']/percentage_city_df['Laid_Off_Count'].sum()*100).round(decimals=2)
percentage_city_df.head()
```

```python
layoffs_data['datee'] = pd.to_datetime(layoffs_data["Date"]).dt.strftime('%Y-%m')
layoffs_data.head()
```

```python
layoffs_data.tail()
```

```python
layoffs_data.dtypes
```

```python
layoffs_data = layoffs_data['Date'].dt.strftime('%m-%Y')
layoffs_data.head()
```

```python
layoffs_data.plot.line(x='datee',y='Laid_Off_Count')
```

```python
len(layoffs_data)
```

```python
clean_layoff_df = layoffs_data.dropna(subset=['Laid_Off_Count'])
len(clean_layoff_df)
```

```python
clean_layoff_df.head()
```

```python
clean_layoff_df[['datee','Laid_Off_Count']].groupby('datee').sum().reset_index().plot.line(x='datee',y='Laid_Off_Count',title="Global Layoff Chart Since COVID-19",xlabel="Date",rot=45)
```

```python
cons_layoff_df = clean_layoff_df.loc[clean_layoff_df['Industry']=='Consumer',:]
cons_layoff_df[['datee','Laid_Off_Count']].groupby('datee').sum().reset_index().plot.line(x='datee',y='Laid_Off_Count',title="Consumer Industry Layoff Chart Since COVID-19",xlabel="Date",rot=45)
```
